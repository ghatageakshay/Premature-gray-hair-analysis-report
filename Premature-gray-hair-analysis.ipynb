```python
import pandas as pd 
import numpy as np


```

```python
df = pd.read_csv('data.csv')


df.head()
```

```python

df.drop(columns = ['Timestamp', '  What is your gender?  ', '  Do you smoke or consume alcohol?  ','Do you eat junk food often?'],inplace=True)

df.columns
```

```python
df.columns = ['Age','Family_history','Gray_onset','stress','sleep','fast_food','screen_time','late_phone']

df.head()
```

```python

df.isna().sum()
```

```python
avg_age = df['Age'].median()
print(avg_age)
df['Age'].fillna(value = avg_age,inplace = True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python

df['Gray_onset']
df.dropna(subset=['Gray_onset'],inplace=True)
```

```python
mean_stress = df['stress'].mean()
df['stress'].fillna(value = mean_stress,inplace=True)
```

```python
#missing values in fast food
fill_fast_food = df['fast_food'].mode()[0]
df['fast_food'].fillna(value = fill_fast_food,inplace=True)

#missing values in screen time
avg_screen_time = df['screen_time'].mode()[0]
df['screen_time'].fillna(value = avg_screen_time,inplace=True)

#missing values in late phone
avg_late_phone = df['late_phone'].mode()[0]
df['late_phone'].fillna(value = avg_late_phone,inplace = True)
```

```python
df.isna().sum()
```

```python
print(df.shape)
print(df.isnull().sum())
df.head()
```

```python
df['Family_history'].value_counts()
```

```python
print(df['Family_history'].unique())
print(df['sleep'].unique())
print(df['late_phone'].unique())
print(df['fast_food'].unique())
print(df['screen_time'].unique())
```

```python
df['Family_history'] = df['Family_history'].map({"yes":1,"no":0})
df['sleep'] = df['sleep'].map({"Yes":2,"Sometimes":1,"No":0})
df['late_phone'] = df['late_phone'].map({"Yes":2,"Sometimes":1,"No":0})
df['fast_food'] = df['fast_food'].map({"Often":2,"Sometimes":1,"Rarely":0})
df['screen_time'] = df['screen_time'].map({"More than 6":4,"4 to 6":3, "2 to 4":2,"Less than 2":1})
```

```python
df.head()
```

```python
print(df.isnull().sum())
print(df.shape)
df.head()
```

```python
X = df.drop(columns=['Gray_onset'])
Y = df['Gray_onset']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,Y_train)

print("Model Trianed")
```

```python
Y_pred = model.predict(X_test)

print("Actual Values: ",Y_test.round(2))
print("Predicted Values: " ,Y_pred.round(2))
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

mae = mean_absolute_error(Y_test, Y_pred)
r2  = r2_score(Y_test, Y_pred)

print(f"Mean Absolute Error: {mae:.2f} years")
print(f"R2 Score: {r2:.2f}")
```
